```python
import pandas as pd

train=pd.read_csv('trainn.csv')
train2=pd.read_csv('train.csv',index_col=0)
```

```python
train.head()
```

```python
final_train=pd.concat([train,train2])
final_train
```

```python
final_train.info()
```

```python
X,Y=final_train.drop('prognosis',axis=1),final_train.prognosis
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

y_enc=le.fit_transform(Y)
```

```python
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')

for i in range(10,100):
    x_train,x_test,y_train,y_test=train_test_split(X,y_enc,test_size=.10,random_state=i)

    from sklearn.linear_model import LogisticRegression

    log=LogisticRegression()
    log.fit(x_train,y_train)
    predict=log.predict(x_test)

    from sklearn.metrics import classification_report
    print(f'Random State: {i}')
    print(classification_report(y_test,predict))
```

```python
from xgboost import XGBClassifier

for i in range(10,100):
    x_train,x_test,y_train,y_test=train_test_split(X,y_enc,test_size=.10,random_state=i)

    xgb=XGBClassifier()
    xgb.fit(x_train,y_train)

    predict=xgb.predict(x_test)
    print(f'Random State: {i}')
    print(classification_report(y_test,predict))
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y_enc,test_size=.10,random_state=67)

xgb=XGBClassifier()
xgb.fit(x_train,y_train)

predict=xgb.predict(x_test)
print(classification_report(y_test,predict))
```

```python
test=pd.read_csv('test.csv',index_col=0)
final_df=pd.DataFrame(test.index)

final_predict=xgb.predict(test)
final_df['prognosis']=le.inverse_transform(final_predict)
final_df.to_csv('submission.csv',index=False)
```
