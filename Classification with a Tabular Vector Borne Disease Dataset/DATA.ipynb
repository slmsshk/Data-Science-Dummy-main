```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y_enc=pd.DataFrame(le.fit_transform(Y),columns=[Y.name])
y_enc
```

```python
from sklearn.manifold import TSNE

tsne=TSNE()
scores=pd.DataFrame(tsne.fit_transform(X),columns=['a','b'])

```

```python
fig, ax = plt.subplots(figsize=(7, 5))

scatter = ax.scatter(scores.a, scores.b, c=y_enc.values)

# produce a legend with the unique colors from the scatter
legend1 = ax.legend(*scatter.legend_elements(),
                    loc="best", title="Classes")
ax.add_artist(legend1)

plt.show()
```

```python
# Class Distribution
height = Y.value_counts().values
x = Y.unique()

fig, ax = plt.subplots(figsize=(15, 7))

# Create a bar plot of the data
bars = ax.bar(x, height, width=0.8)

# Add axis labels and a title
ax.set_xlabel('Classes')
ax.set_ylabel('Count')
ax.set_title('Class Distribution')

# Add grid lines
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Add total count on top of each bar
for i, bar in enumerate(bars):
    ax.text(bar.get_x() + bar.get_width() / 2, height[i] , height[i],
            ha='center', va='bottom', color='green', fontsize=12, fontweight='bold')

# Color each bar differently
colors = plt.cm.Set3(np.linspace(0, 1, len(x)))
for i, bar in enumerate(bars):
    bar.set_color(colors[i])

# Show the plot
plt.show()
```

```python
from sklearn.decomposition import KernelPCA

kpca=KernelPCA(2)
pcs=pd.DataFrame(kpca.fit_transform(X),columns=['a','b'])

plt.scatter(pcs.a,pcs.b,c=y_enc.values)
plt.show()
```

```python
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.decomposition import KernelPCA

for i in range(20,100):
    est=[]

#     est.append(('Scaler',MinMaxScaler()))
    # est.append(('PCA',KernelPCA(2)))
    est.append(('logistic_regression',LogisticRegression()))
    pipe=Pipeline(est)

    x_train,x_test,y_train,y_test=train_test_split(X,y_enc,test_size=.10,random_state=i)
    pipe.fit(x_train,y_train)
    pred=pipe.predict(x_test)
    print(f'Classification Report: random state{i}')
    print(classification_report(y_test,pred))
```

```python

est=[]

est.append(('Scaler',MinMaxScaler()))
# est.append(('PCA',KernelPCA(2)))
est.append(('logistic_regression',LogisticRegression()))
pipe=Pipeline(est)

x_train,x_test,y_train,y_test=train_test_split(X,y_enc,test_size=.10,random_state=23)
pipe.fit(x_train,y_train)
pred=pipe.predict(x_test)
print(f'Classification Report:')
print(classification_report(y_test,pred))
```

```python
test=pd.read_csv('test.csv',index_col=0)
final_df=pd.DataFrame(test.index)

final_predict=pipe.predict(test)
final_df['prognosis']=le.inverse_transform(final_predict)
final_df.to_csv('submission.csv',index=False)
```
