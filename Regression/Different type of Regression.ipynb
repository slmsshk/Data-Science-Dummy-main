# SLR

```python
list(data.columns).remove('AgeInDays')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

def rmse(y,pred):
    return np.sqrt(mean_squared_error(y,pred))

lr=LinearRegression()
s={}

Y=data.Strength
for i in data.columns[:-1]:
    for j in data.columns[:-1]:
        if i==j:
            continue
        else:
            X=data.drop([i,'Strength'],axis=1)
            lr.fit(X,Y)
            pred=lr.predict(X)
            s[f'leaving {i}']=lr.score(X,Y)
            s[f'mse leaving {i}']=rmse(Y,pred)

s
```

# Quantile Regression

```python
import statsmodels.api as sm
# Specify the quantile(s) of interest
quantiles = [0.25, 0.5, 0.75]

# Fit the quantile regression model for each quantile
models = []
for q in quantiles:
    model = sm.QuantReg(Y, X).fit(q=q)
    models.append(model)

# Print the summary of each model
for i, model in enumerate(models):
    print(f"Quantile regression model for {quantiles[i]}:")
    print(model.summary())
```

```python
for i in models:
    print(rmse(Y,i.predict(X)))
```

```python
final_dict['Quantile regressor']=models[1].predict(X)
test_dict['Quantile regressor']=models[1].predict(test)
```

```python
final_dict,test_dict
```
