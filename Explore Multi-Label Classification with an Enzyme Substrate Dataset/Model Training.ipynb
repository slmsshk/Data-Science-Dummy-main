```python
import pandas as pd

data = pd.read_csv('train.csv',index_col=0)
data.head()
```

```python
data
```

```python
y_cols = list(data.columns).index('fr_COO')
X,Y = data.iloc[:,:y_cols],data[data.columns[y_cols:]]
X.head(),Y.head()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=.20)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.multioutput import MultiOutputClassifier

log = MultiOutputClassifier(SVC(kernel='linear'))

log.fit(x_train,y_train)
```

```python
pred=log.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, multilabel_confusion_matrix

# Calculate evaluation metrics
# accuracy = accuracy_score(y_test, pred)
precision = precision_score(y_test, pred, average='micro')
recall = recall_score(y_test, pred, average='micro')
f1 = f1_score(y_test, pred, average='micro')

# Create multilabel confusion matrix
confusion_mat = multilabel_confusion_matrix(y_test, pred)

# Print evaluation metrics
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)

# Print confusion matrix for each label
for i, cm in enumerate(confusion_mat):
    print(f"\nConfusion Matrix for Label {i+1}:")
    print(cm)
```
