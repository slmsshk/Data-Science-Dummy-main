# Final Model

```python
import pandas as pd
```

```python
train=pd.read_csv('train.csv')
train
```

```python
X,Y=train.iloc[:,1:-1],train.stroke
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=.20)
```

```python
from sklearn.pipeline import Pipeline

from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.decomposition import PCA

from xgboost import XGBClassifier
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier,RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression,SGDClassifier
from sklearn.neural_network import MLPClassifier

from xgboost import XGBClassifier
```

```python
# Stacked
mod=[]

dtc=DecisionTreeClassifier()
mod.append(('DTC',dtc))

ada=AdaBoostClassifier()
mod.append(('ada',ada))

log=LogisticRegression()
mod.append(('log',log))

forest=RandomForestClassifier()
mod.append(('forest',forest))

svc=SVC()
mod.append(('svc',svc))

knn=KNeighborsClassifier()
mod.append(('knn',knn))

nn=MLPClassifier()
mod.append(('nn',nn))

sgd=SGDClassifier()
mod.append(('sgd',sgd))

xgb=XGBClassifier()
mod.append(('xgb',xgb))

vote=VotingClassifier(mod)
vote
```

```python
est=[]

ohe=OneHotEncoder(sparse=False,handle_unknown='ignore')
est.append(('Encoder',ohe))

scaler=StandardScaler()
est.append(('scaler',scaler))

pca=PCA(15)
est.append(('pca',pca))

# xgb=XGBClassifier()
# est.append(('model',xgb))

est.append(('Voting',vote))

pipe=Pipeline(est)
pipe
```

```python
pipe.fit(x_train,y_train)
```

```python
predict=pipe.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,predict))
```

```python
test_X=pd.read_csv('test.csv').drop('id',axis=1)
pred_test=pipe.predict(test_X)
```
